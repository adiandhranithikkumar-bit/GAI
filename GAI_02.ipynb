```python
import pytesseract
from PIL import Image
from transformers import pipeline
image_path = '/content/picture gen ai.webp'
image = Image.open(image_path)
extracted_text = pytesseract.image_to_string(image).strip()
if not extracted_text:
    print("No text extracted from the image.")
else:
    print(f"Extracted Test: {extracted_text}")
    sentiment_analysis = pipeline("sentiment-analysis")
    sentiment_result = sentiment_analysis(extracted_text)[0]
    print("\nSentiment Analysis:")
    print(f"Text: {extracted_text}")
    print(f"Sentiment: {sentiment_result['label']}, (Score: {sentiment_result['score']})")
```
